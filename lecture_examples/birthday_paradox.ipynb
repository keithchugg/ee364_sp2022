# Brithday Paradox Notebook
The birdthday paradox concerns the probability that a room/class with $k$ people will have at least one birthday "clash" -- i.e., the probability that at least two people in the room will share a common birdthday (not condering year).  Deonte this event as $A$. 

The easiest way to compute this probability is to consider the compliment probability: the event that none of the `k` people share a common birthday and use $P(A) = 1 - P(A^c)$

To find the probability of $A^c$, we should count the number of ways that birthdays can be assigned without a clash and the number of ways that birthdays can be assigned without regard to clashes:

$P(A^c) = N_{\mathrm{clash-free}} / N_{\mathrm{anyway}}$

where $N_{\mathrm{clash-free}}$ is the number of ways to assign b-days without a clash and $N_{\mathrm{anyway}}$ is the number of ways to assign b-days without regard to clashes. 

It follows that 

$P(A^c) = \frac{(365)(364)(363) \cdots (365-k-1)}{(365)(365)(365) ]cdots (365)} = \frac{(365)_k}{365^k}$

To make the numerical computation simple, we can write the probability of no common b-days as

$P(A) = 1 - \prod_{m=0}^{k-1} \frac{365-m}{365}$

Let's code this up and plot the probability of a common b-day as a function of the class size.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# this does this is a numerically stable manner
def p_clash(k):
    p = 1
    for i in range(k):
        p = p * (365 - i) / 365
    return 1 - p
```

Now let's create an array of the values of `p_clash` for different class sizes:

```python
N_students_max = 60

p_clashes = np.zeros(N_students_max)
for k in range(N_students_max):
    p_clashes[k] = p_clash(k + 1)
```

and plot the results.  

```python
plt.figure(figsize = (10,10))
plt.plot(np.arange(N_students_max) + 1, p_clashes)
plt.axhline(y = 0.5, c = 'r', linestyle = '--')
plt.axvline(x=23, c='r', linestyle = '--')
plt.title('Birthday Paradox: with 23 or more people, clashed more probable than not')
ax = plt.gca()
ax.set_xlabel('class size (k)')
ax.set_ylabel('probability common birthday')
```

Note that for classes of size 23 or larger, common birthdays are more common than not.  This is the paradox: the intuition for most people is that one would need very large class size before common birthdays are probable!
